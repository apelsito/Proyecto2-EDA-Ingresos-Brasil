# Fase 4 Visualización

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sys
sys.path.append("../") #Para poder importar
warnings.filterwarnings("ignore")
pd.options.display.float_format = "{:.2f}".format
# Carga el dataset de trabajo
pd.set_option("display.max_columns", None) #Ver todo
df = pd.read_csv("../datos/output/datos_2013_to_2021_Final.csv", delimiter=",",index_col=0)
```

# Ingresos previstos vs registrados vs ejecutados por categoría

```python
cond1 = df["categoría económica"] == "Receitas Correntes"
cond2 = df["categoría económica"] == "Receitas Correntes - intra-orçamentárias"
cond3 = df["categoría económica"] == "Receitas de Capital"
cond4 = df["categoría económica"] == "Receitas de Capital - intra-orçamentárias"
cond5 = df["categoría económica"] == "Sem informação"

ca1 = df[cond1]
ca2 = df[cond2]
ca3 = df[cond3]
ca4 = df[cond4]
ca5 = df[cond5]

categ1 = ca1.groupby(["categoría económica", "año fiscal"]).agg({"valor previsto actualizado": "sum","valor registrado": "sum","valor ejecutado": "sum"}).reset_index()
categ2 = ca2.groupby(["categoría económica", "año fiscal"]).agg({"valor previsto actualizado": "sum","valor registrado": "sum","valor ejecutado": "sum"}).reset_index()
categ3 = ca3.groupby(["categoría económica", "año fiscal"]).agg({"valor previsto actualizado": "sum","valor registrado": "sum","valor ejecutado": "sum"}).reset_index()
categ4 = ca4.groupby(["categoría económica", "año fiscal"]).agg({"valor previsto actualizado": "sum","valor registrado": "sum","valor ejecutado": "sum"}).reset_index()
categ5 = ca5.groupby(["categoría económica", "año fiscal"]).agg({"valor previsto actualizado": "sum","valor registrado": "sum","valor ejecutado": "sum"}).reset_index()

fig, axes = plt.subplots(3,2, figsize=(16, 12)) 
axes = axes.flatten()
sns.lineplot(x="año fiscal", y="valor previsto actualizado", data=categ1, marker="o",label="valor previsto actualizado",ax=axes[0])
sns.lineplot(x="año fiscal", y="valor registrado", data=categ1, marker="o", label="valor registrado", linestyle="--",ax=axes[0])
sns.lineplot(x="año fiscal", y="valor ejecutado", data=categ1, marker="o",label="valor ejecutado" , linestyle="-.",ax=axes[0])
axes[0].set_title("Ingresos previstos, vs Registrado, vs ejecutados en Receitas Correntes")
axes[0].legend(fontsize=8, loc='upper left')
axes[0].set_xlabel("Año Fiscal")
axes[0].set_ylabel("valores (En Millones)")
axes[0].set_xticks(ticks=[2013,2014,2015,2016,2017,2018,2019,2020,2021])
axes[0].grid()
sns.lineplot(x="año fiscal", y="valor previsto actualizado", data=categ2, marker="o",label="valor previsto actualizado",ax=axes[1])
sns.lineplot(x="año fiscal", y="valor registrado", data=categ2, marker="o", label="valor registrado", linestyle="--",ax=axes[1])
sns.lineplot(x="año fiscal", y="valor ejecutado", data=categ2, marker="o",label="valor ejecutado" , linestyle="-.",ax=axes[1])
axes[1].set_title("Ingresos previstos, vs Registrado, vs ejecutados en Receitas Correntes - intra-orçamentárias")
axes[1].legend(fontsize=8, loc='upper left')
axes[1].set_xlabel("Año Fiscal")
axes[1].set_ylabel("valores (En Millones)")
axes[1].set_xticks(ticks=[2013,2014,2015,2016,2017,2018,2019,2020,2021])
axes[1].grid()
sns.lineplot(x="año fiscal", y="valor previsto actualizado", data=categ3, marker="o",label="valor previsto actualizado",ax=axes[2])
sns.lineplot(x="año fiscal", y="valor registrado", data=categ3, marker="o", label="valor registrado", linestyle="--",ax=axes[2])
sns.lineplot(x="año fiscal", y="valor ejecutado", data=categ3, marker="o",label="valor ejecutado" , linestyle="-.",ax=axes[2])
axes[2].set_title("Ingresos previstos, vs Registrado, vs ejecutados en Receitas de Capital")
axes[2].legend(fontsize=8, loc='upper left')
axes[2].set_xlabel("Año Fiscal")
axes[2].set_ylabel("valores (En Millones)")
axes[2].set_xticks(ticks=[2013,2014,2015,2016,2017,2018,2019,2020,2021])
axes[2].grid()
sns.lineplot(x="año fiscal", y="valor previsto actualizado", data=categ4, marker="o",label="valor previsto actualizado",ax=axes[3])
sns.lineplot(x="año fiscal", y="valor registrado", data=categ4, marker="o", label="valor registrado", linestyle="--",ax=axes[3])
sns.lineplot(x="año fiscal", y="valor ejecutado", data=categ4, marker="o",label="valor ejecutado" , linestyle="-.",ax=axes[3])
axes[3].set_title("Ingresos previstos, vs Registrado, vs ejecutados en Receitas de Capital - intra-orçamentárias")
axes[3].legend(fontsize=8, loc='upper left')
axes[3].set_xlabel("Año Fiscal")
axes[3].set_ylabel("valores (En Millones)")
axes[3].set_xticks(ticks=[2013,2014,2015,2016,2017,2018,2019,2020,2021])
axes[3].grid()
sns.lineplot(x="año fiscal", y="valor previsto actualizado", data=categ5, marker="o",label="valor previsto actualizado",ax=axes[4])
sns.lineplot(x="año fiscal", y="valor registrado", data=categ5, marker="o", label="valor registrado", linestyle="--",ax=axes[4])
sns.lineplot(x="año fiscal", y="valor ejecutado", data=categ5, marker="o",label="valor ejecutado" , linestyle="-.",ax=axes[4])
axes[4].set_title("Ingresos previstos, vs Registrado, vs ejecutados Categoría sin información")
axes[4].legend(fontsize=8, loc='upper left')
axes[4].set_xlabel("Año Fiscal")
axes[4].set_ylabel("valores (En Millones)")
axes[4].set_xticks(ticks=[2013,2014,2015,2016,2017,2018,2019,2020,2021])
axes[4].grid()
plt.tight_layout()
```

# Evolución Temporal de ingresos previstos vs registrado vs ejecutados

```python
categ1 = df.groupby("año fiscal")["valor previsto actualizado"].sum().reset_index()
categ2 = df.groupby("año fiscal")["valor registrado"].sum().reset_index()
categ3 = df.groupby("año fiscal")["valor ejecutado"].sum().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(x="año fiscal", y="valor previsto actualizado", data=categ1, marker="o",label="valor previsto actualizado")
sns.lineplot(x="año fiscal", y="valor registrado", data=categ2, marker="o", label="valor registrado", linestyle="--")
sns.lineplot(x="año fiscal", y="valor ejecutado", data=categ3, marker="o",label="valor ejecutado" , linestyle="-.")
plt.title("Ingresos previstos, vs Registrado, vs ejecutados en Receitas Correntes")
plt.legend(fontsize=10, loc='upper left')
plt.xlabel("Año Fiscal")
plt.ylabel("valores (En Millones)")
plt.xticks(ticks=[2013,2014,2015,2016,2017,2018,2019,2020,2021])
plt.grid()
plt.tight_layout()
```
